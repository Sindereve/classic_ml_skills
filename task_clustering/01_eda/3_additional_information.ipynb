# Дополнительная информация

## Метрики (личная боль для этой задачи)

Прочесав и прочтя большое количество материалов из интернета можно выделить несколько метрик:
* метрики качества кластеризации; \[[habr](https://habr.com/ru/companies/yandex/articles/500742/),[labex](https://labex.io/ru/tutorials/adjusting-for-chance-in-clustering-performance-evaluation-49059)\]
* бизнес-интерпретации класстеров;
* визуализация.

*Визуализацию мы не будем разбирать*, так как подобранный нами датасет сложен в интерпритации его на 2D полотно. Большинство реальных данных тяжело разделить использую 2D или 3D график. (Пробовал изменить датасет с использованием PCA, но обьяснимость дисперсии равнялася 0.17)

### Метрики качества кластеров

|Метрика|Описание|Диапазон значений|Заметки|
|-|-|-|-|
|Silhouette Score|Измеряет на сколько хорошо разделяются класстеры. Реагирует на пересечения класстеров|-1..1|1 - Идеальное разделение, 0 - Кластеры пересекаются, -1 - Лбъекты назначены неверным кластерам|
|Calinski-Harabasz Score|Отношение межкластерной дисперсии к внутрикластерной|0...|Чем больше тем лучше|
|Davies-Bouldin Score|Среднее сходство между каждым кластером и его наиболее похожим кластером.|0...|Чем ближе к 0 тем лучше|

Получать данные мы сможем двумя способами:
- Метод ```to_dataframe``` - получаем информацию о всех наших моделях в формате ```pd.DataFrame```;
- Индексирование ```history[index]```, где ```history``` - объект нашего класса. На выходе мы получим обученную модель под заданным индексом. Принцип работы аналогичен методу ```DataFrame.iloc[index]```.
