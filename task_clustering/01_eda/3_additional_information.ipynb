## Метрики (личная боль для этой задачи)

Прочесав и прочтя большое количество материалов из интернета можно выделить несколько метрик:
* метрики качества кластеризации \[[habr](https://habr.com/ru/companies/yandex/articles/500742/),[labex](https://labex.io/ru/tutorials/adjusting-for-chance-in-clustering-performance-evaluation-49059)\];
* бизнес-интерпретации класстеров \[[sky](https://sky.pro/wiki/analytics/metod-klasterov-printsipy-analiza-i-sfery-primeneniya-v-analitike/)\];
* визуализация.

*Визуализацию мы не будем разбирать*, так как подобранный нами датасет сложен в интерпритации его на 2D полотно. Большинство реальных данных тяжело разделить использую 2D или 3D график. (Пробовал изменить датасет с использованием PCA, но обьяснимость дисперсии равнялася 0.17)

### Бизнес-интерпретация кластеров

Бизнес-метрики - это качествыенные показатели, которые оценивает не математическое качество кластеров, а их практическую полезность для бизнеса.

В контексте автомобилей BMW бизнес-метрики позволяют перевести абстрактные математические кластеры в конкретные рыночные сегменты, которые можно использовать для принятия коммерческих решений. В отличие от математических метрик, которые оценивают формальное качество разделения данных, бизнес-метрики отвечают на вопрос - насколько выделенные кластеры полезны для реальной работы с автомобилями: закупки, ценообразования, маркетинга и продаж.

Для нашего датасета BMW ключевыми бизнес-метриками являются интерпретируемость кластеров - можем ли мы дать каждому кластеру осмысленное названия:
* бюджетные седаны для города;
* премиальные внедорожники;
* спортивные модели. 
 
Вторая важная метрика - сбалансированность сегментов, чтобы каждый кластер был достаточно большим для целевого маркетинга. 

Третья метрика - профилируемость, то есть возможность четко описать типичного представителя каждого кластера по году выпуска, цене, пробегу и другим характеристикам. 

Четвертая метрика - действенность, которая показывает, можно ли на основе кластеризации принимать конкретные бизнес-решения, такие как оптимизация ценовой политики для разных сегментов или корректировка ассортимента.

Именно бизнес-метрики превращают кластеризацию из академического упражнения в практический инструмент управления автомобильным бизнесом, позволяя выделять целевые аудитории, оптимизировать рекламные бюджеты и повышать эффективность продаж за счет лучшего понимания структуры рынка.
